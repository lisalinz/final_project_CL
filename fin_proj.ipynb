```python
# избавимся от шума
from nltk.corpus import stopwords
stop = stopwords.words('english')
pat = r'\d+'.format('|'.join(stop))
pat1=r'[^\w\s]'.format('|'.join(stop))
pat2=r'\n'

all_revs=pd.read_csv('all_rw_shuf.csv')
all_revs['reviews_clean'] = all_revs['reviews'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
all_revs['reviews_clean'] = all_revs['reviews_clean'].str.replace(pat, '')
all_revs['reviews_clean'] = all_revs['reviews_clean'].str.replace(pat1, '')
all_revs['reviews_clean'] = all_revs['reviews_clean'].str.replace(pat2, '')
all_revs['reviews_clean'] = all_revs['reviews_clean'].str.lower()

print(all_revs)
```

```python
all_revs = all_revs.drop(['reviews'],axis=1)
all_revs = all_revs.drop(['Unnamed: 0'],axis=1)

display(all_revs)
```

```python
all_revs.to_csv('all_revs_clean.csv')
```

```python
# начинаем применять модели
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# сбалансируем классы
revs_balanced = all_revs[all_revs.v1 == 1].append(all_revs[all_revs.v1 == 0].sample(n=747))
revs_balanced.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(revs_balanced['reviews_clean'])
```

```python
# разделим на обучающую и тестовую выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, revs_balanced["v1"], test_size=0.3, random_state=25)
print(X_train.shape, X_test.shape)
```

```python
# БЭГГИНГ
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
```

```python
tree_y_pred = tree_clf.predict(X_test)
tree_y_pred
```

```python
# сравним, насколько предсказания нашего дерева совпали с настоящими лейблами
print(classification_report(y_test, tree_y_pred))
```

```python
# обучим Random Forest
from sklearn.ensemble import RandomForestClassifier 
forest = RandomForestClassifier(n_estimators=500, verbose=1)
```

```python
forest.fit(X_train, y_train)

forest_y_pred = forest.predict(X_test)

print(classification_report(y_test, forest_y_pred))
```

```python
# БУСТИНГ
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=1000, random_state=10, verbose=1)
```

```python
gb_clf.fit(X_train, y_train)
```

```python
gb_y_pred = gb_clf.predict(X_test)
```

```python
# проверяем

print(classification_report(y_test, gb_y_pred))

# кажестя, сработал хуже, чем бэггинг (просел precision)
```

```python
# обучим классификатор

gb_clf_es = GradientBoostingClassifier(n_iter_no_change=5, verbose=10)

gb_clf_es.fit(X_train, y_train)
```

```python
es_y_pred = gb_clf_es.predict(X_test)
```

```python
print(classification_report(y_test, es_y_pred))
```

```python
# определим уникальные слова для отзывов
revs_pop=pd.DataFrame(tree_clf.feature_importances_)

revs_pop["feature"] = vectorizer.get_feature_names()
```

```python
revs_pop.nlargest(10,0)

# сравни алгоритмы, чтобы определить тот, который сработал лучше всего
```

```python
# Random Forest
revs_pop2 = pd.DataFrame(forest.feature_importances_)

revs_pop2["feature"] = vectorizer.get_feature_names()

revs_pop2.nlargest(10,0)
```

```python
# GradientBoostingClassifier
revs_pop3 = pd.DataFrame(gb_clf.feature_importances_)

revs_pop3["feature"] = vectorizer.get_feature_names()

revs_pop3.nlargest(10,0)
```

```python
# GradientBoostingClassifier + Early stopping

revs_pop4 = pd.DataFrame(gb_clf_es.feature_importances_)

revs_pop4["feature"] = vectorizer.get_feature_names()

revs_pop4.nlargest(10,0)
```
